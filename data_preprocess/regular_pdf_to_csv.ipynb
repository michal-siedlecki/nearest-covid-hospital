```python
import tabula
import pandas as pd
import os
import re
```

```python
os.chdir(os.path.join(os.getcwd(),'data_pdf/pdf/'))
```

```python
pdfs = [x for x in os.listdir() if x.endswith('.pdf')]
```

```python
table_list = []

for pdf in pdfs:
    tables = tabula.read_pdf(pdf, multiple_tables=True, pages='all')
    for table in tables:
        table_list.append([table, pdf])
```

```python
df_2_col, df_3_col, df_4_col = [], [], []

for l in table_list:
    col_num = len(l[0].columns)
    if col_num == 2:
        df_2_col.append(l[0])
    elif col_num == 3:
        df_3_col.append(l[0])
    else:
        df_4_col.append(l[0])
            
```

```python
for frame in df_2_col:
    frame.columns = [1, 2]
    
for frame in df_3_col:
    frame.columns = [1, 2, 3]
    
for frame in df_4_col:
    frame.columns = [1, 2, 4, 5]
```

```python
df = pd.concat(df_2_col)
df.reset_index(drop=True, inplace=True)
df_3 = pd.concat(df_3_col)
df_3.reset_index(drop=True, inplace=True)
df_3.fillna('', inplace=True)
df_3.head()
```

```python
df_3[2] = df_3[2] + df_3[3]
df_3.drop(columns=[3], inplace=True)
df = df.append(df_3)
df.reset_index(drop=True, inplace=True)
df.tail(10)
```

```python
df_4 = pd.concat(df_4_col)
df_4.reset_index(drop=True, inplace=True)
df_4.fillna('', inplace=True)
df_4.head()
```

```python
df_4[1] = df_4[1] + df_4[2]
df_4.drop(columns=[2], inplace=True)
df_4[4] = df_4[4] + df_4[5]
df_4.drop(columns=[5], inplace=True)
```
